```python
# READING THE DATA FROM INVENTORY :

fd = open ("Inventory_management_system.txt","r")  # open a file
product = fd.read().split("\n")                    # read this file 
for prod in product:
    print(prod)                                    # printing the Products
fd.close()                                         # close the file 

# TAKING USER_DETAILS :

user_Name = input("Enter Your Name : ")
user_Email = input ("Enter Email Id : ")
user_ 
u_prod_id = input("Enter Product Id : ")           # get product id from user 




# GOING THROUGH EACH PRODUCT DETAILS :

updated_prod_list = []
lst = []

for prod in product:
    prod_details = prod.split(' : ')               # searching that the product is available or not?
    if (u_prod_id == prod_details[0]):                     # if yes then print the product_details
        print("-------------------------------------")
        print ("Product Name : " , prod_details[1])
        print ("Price : " , prod_details[2])
        print ("Product available : " , prod_details[3])
        u_prod_qn = input("Enter Product Quantity : ")         # get how much quantity user need
        print("-------------------------------------")
        
# CHECKING PRODUCT IS AVAILABLE OR NOT :

        if (int(prod_details[3]) == 0):
            print("Sorry Product is Out Of Stock ...")             # Product is not available 
        elif (int(u_prod_qn) <= int(prod_details[3])):            # check product availability (if yes then proceed)
            prod_details[3] = str(int(prod_details[3]) - int(u_prod_qn))   # after buy new_product_quantity
            print ("Quantity : " , u_prod_qn)                              # print Quantity
            Amount =  (int(u_prod_qn) * int(prod_details[2]))              # print Amount
            print("-------------------------------------")
            print("Billing Amount : ",  Amount)                            # print Bill
            print("-------------------------------------")  
        else:
            print("-------------------------------------")
            print("Only " + str(prod_details[3])  + " Product_Quantity is Available")    # product available but not that much 
            print ("Do You Want To Buy All ? ")                        
            u_opinion = input("Enter yes or no ...   ")
            if str(u_opinion) == "yes" :
                Amount =  (int(prod_details[3]) * int(prod_details[2]))              # print Amount
                print("-------------------------------------")
                print("Billing Amount : ",  Amount)                            # print Bill
                print("-------------------------------------")
                prod_details[3] = "0"   
            else :
                print("Thanks For Comming ... ")              
            
# UPDATED PRODUCT LIST:

    updated_prod_list.append(prod_details)    # After buying Update_new_Product_details_list

for i in updated_prod_list:
    u_prod = i[0] + " : " + i[1] + " : " + i[2] + " : " + i[3] + "\n"      # product_list in this formate
    print(u_prod)              # Print Updated Product_list
    lst.append(u_prod)
    
lst[-1] = lst[-1][:-1]      # after last product_details enter, don't generate any new line *otherwise__error*

fd = open ("Inventory_management_system.txt","w") 
for i in lst:
    fd.write(i)                                        # Override the file after buying a product
fd.close()
```
